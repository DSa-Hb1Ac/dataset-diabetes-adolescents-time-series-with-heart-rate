```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data-csv/Population.csv')
df.head(3)
```

```python
# correlation heatmap
corr = df.drop(columns=["HbA1C_1 [yyyy]"]).corr()
sns.heatmap(corr,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            annot=True)
plt.show()
```

```python
sns.countplot(data=df, x="Gender [M=male F=female]")
plt.show()
```

```python
sns.countplot(data=df, x="Pump brand")
plt.show()
```

```python
sns.violinplot(data=df, y="Length [cm]", x="Gender [M=male F=female]")
plt.show()
```

```python
sns.countplot(data=df, x="Age [yr]", hue="Gender [M=male F=female]")
plt.show()
```
